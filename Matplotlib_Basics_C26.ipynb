```python
# some complex plots
x_val = [-3,-2,-1,0,1,2,3]
y_val = [9,4,1,0,1,4,9]
plt.plot(x_val,y_val)
plt.show()
```

```python
# using numpy
x_val = np.array([-3,-2,-1,0,1,2,3])
y_val = x_val**2
plt.plot(x_val,y_val)
plt.show()
```

```python
x_val = np.arange(-4,4,.1)
y_val = x_val**2
plt.plot(x_val,y_val)
plt.show()
```

```python
x_val = np.arange(-4,4,.1)
y_val = x_val**2
plt.xticks(np.arange(-4,4,.3),rotation=90)
plt.plot(x_val,y_val)
plt.show()
```
