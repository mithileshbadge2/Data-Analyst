```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('Sales_Funnel_CRM.csv')
```

```python
df
```

```python
df.shape
```

```python
Licenses = df[['Company','Product','Licenses']]
```

```python
Licenses
```

```python
pd.pivot(data=Licenses,index='Company',columns='Product',values='Licenses')
```

```python
# using pivot table
pd.pivot_table(data=Licenses,index='Company')
```

```python
pd.pivot_table(data=df,index='Company')
```

```python
pd.pivot_table(data=Licenses,index='Company',aggfunc='sum')
```

```python
df.groupby('Company').sum()
```

```python
pd.pivot_table(data=df,index='Company',aggfunc='sum')
```

```python
pd.pivot_table(data=df,index='Company',values=['Licenses','Sale Price'], aggfunc='sum')
```

```python
pd.pivot_table(data=df,index='Account Manager',values=['Sale Price'],aggfunc='sum')
```

```python
pd.pivot_table(data=df,index=['Account Manager','Contact'],values=['Sale Price'],aggfunc='sum')
```

```python
pd.pivot_table(data=df,index=['Account Manager','Contact'],values=['Sale Price'],columns='Product',aggfunc='sum')
```

```python
# replacing Nan with zero
pd.pivot_table(data=df,index=['Account Manager','Contact'],values=['Sale Price'],columns='Product',aggfunc='sum',fill_value=0)
```

```python
pd.pivot_table(data=df,index=['Account Manager','Contact'],values=['Sale Price'],columns='Product',aggfunc='sum',fill_value=0,margins=True)
```

```python
# rearranging the index and columns for more readability
pd.pivot_table(data=df,index=['Account Manager','Contact','Product'],values=['Sale Price'],aggfunc='sum',fill_value=0,margins=True)
```
