```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# using numpy to create  a cube function graph
x_val = np.arange(-10,10,.5)
y_val = x_val**3
plt.xticks(np.arange(-10,10,1),rotation=90)
plt.plot(x_val,y_val)
plt.show()
```

```python
x_val = np.arange(-4,4,.1)
y_val = x_val**2
plt.xticks(np.arange(-4,4,.3),rotation=90)
plt.plot(x_val,y_val)
plt.show()
```

```python
# ENRICHING THE GRAPH
# changing the color
# changing the default from line to *
x_val = np.arange(-4,4,.1)  #elementwise operation possible
y_val = x_val**2
plt.xticks(np.arange(-4,4,.3),rotation=90)
plt.plot(x_val,y_val,color='red')
plt.show()
```

```python
x_val = np.arange(-4,4,.1)  #elementwise operation possible
y_val = x_val**2
plt.xticks(np.arange(-4,4,.3),rotation=90)
plt.plot(x_val,y_val,color='red',marker='*')
plt.show()
```

```python
x_val = np.arange(-4,4,.1)  #elementwise operation possible
y_val = x_val**2
plt.xticks(np.arange(-4,4,.3),rotation=90)
plt.plot(x_val,y_val,'y*') #yellow star
plt.show()
```

```python
x_val = np.arange(-4,4,.1)  #elementwise operation possible
y_val = x_val**2
plt.xticks(np.arange(-4,4,.3),rotation=90)
plt.plot(x_val,y_val,'b+') #blue star
plt.show()
```

```python
# multiple plots in the same canvas
x_val = np.arange(-4,4,.1)
plt.xticks(np.arange(-4,4,.3),rotation=90)
y1_val = x_val**2
plt.plot(x_val,y1_val)
y2_val = x_val**3
plt.plot(x_val,y2_val)
plt.show()
```

```python
# adding legends
x_val = np.arange(-4,4,.1)
plt.xticks(np.arange(-4,4,.3),rotation=90)
y1_val = x_val**2
y2_val = x_val**3
plt.plot(x_val,y1_val,label="Square")
plt.plot(x_val,y2_val,label="Cube")
plt.legend()
plt.show()
```

```python
x_val = np.arange(-4,4,.1)
plt.xticks(np.arange(-4,4,.3),rotation=90)
y1_val = x_val**2
y2_val = x_val**3
plt.plot(x_val,y1_val,)
plt.plot(x_val,y2_val,)
plt.legend(['Square','Cube'],loc='lower right')
plt.show()
```

```python
x_val = np.arange(-4,4,.1)
plt.xticks(np.arange(-4,4,.3),rotation=90)
y1_val = x_val**2
y2_val = x_val**3
plt.plot(x_val,y1_val,)
plt.plot(x_val,y2_val,)
plt.legend(['Square','Cube'],loc='lower right',frameon=False)
plt.show()
```

```python
x_val = np.arange(-4,4,.1)
plt.xticks(np.arange(-4,4,.3),rotation=90)
y1_val = x_val**2
y2_val = x_val**3
plt.plot(x_val,y1_val,)
plt.plot(x_val,y2_val,)
plt.legend(['Square','Cube'],loc='lower right',frameon=False,ncol=2)
plt.show()
```

```python
x_val = np.arange(-4,4,.1)
plt.xticks(np.arange(-4,4,.3),rotation=90)
y1_val = x_val**2
y2_val = x_val**3
plt.plot(x_val,y1_val,)
plt.plot(x_val,y2_val,)
plt.legend(['Square','Cube'],loc='lower right',frameon=False,ncol=2)
plt.text(0,0+.5,'(0,0)')
plt.text(1,1+.5,'two')
plt.xlim(-2,2)
plt.ylim(-2,2)
plt.show()
```
